Cuando estás trabajando con grandes cantidades de datos en datascience debes intentar encontrar patrones

## Bebés por cada mujer en India

Por ejemplo, esta es una lista de cuantos bebés nacen por cada mujer en India:

| Año  | # bebés    |
| ---- | ---------- |
| 1960 | 5.91       |
| 1970 | 5.59       |
| 1980 | 4.83       |
| 1990 | 4.05       |
| 2000 | 3.31       |
| 2010 | 2.60       |
	
Como es el patrón de datos en la tabla anterior? La mujer está pariendo mas o menos bebés cada año?

### Mira la tabla

En la tabla podemos observar que el número está disminuyendo año tras año, pero pudiese ser una buena idea dibujar una gráfica para poder detectar los patrones mas rápidamente.

Si colocamos estos datos en un gráfico, sería algo así:

El método mas popular de dibujar o graficar funciones en python es utilizando la librería matplotlib, específicamente utilizando el módulo pyplot. Vamos a importar la librería:

````python
# importa el módulo python para plotear
import matplotlib.pyplot as plt
````

Ahora tenemos que convertir la tabla en arrays y graficar los valores:

````python
x_axis = [1960, 1970, 1980, 1990, 2000, 2010]
y_axis = [5.91, 5.59, 4.83, 4.05, 3.31, 2.60]

plt.figure(num=0, dpi=120)
plt.plot(x_axis, y_axis)
````

### Algunas deducciones rápidas?

Al ver los datos uno puede deducir que tenemos una reducción **constante** de nacimiento de niños.
Esta es una función que representa una reducción constante:

```py
def f(x):
    return x - 1
```

y así se ve graficado:


````python
def f(x):
    return x - 1

plt.plot(x_axis, [f(y) for y in range(7,1,-1)])
````

Si trazamos ambos gráficos a la vez, puedes ver similitudes de comportamiento.

````python
plt.plot(x_axis, y_axis)

def f(x):
    return x

plt.plot(x_axis, [f(y) for y in range(7,1,-1)])
````
