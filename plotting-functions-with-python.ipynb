# Plot functions with Python

Plotting data in Python is very similar to the process you do in linear algebra: Given a sample of X values, you use a function to obtain the Y values, and plot the intersection of the X and Y coordinates in the cartesian plane.

In Python we use:

1. The NumPy package for obtaining the X and Y values.
2. The Pyplot module of the Matplotlib library for plotting and drawing the function.

```python
# import the python module to plot
import matplotlib.pyplot as plt
import numpy as np
```

If you don't have real data to plot, you can use NumPy to generate some sample values, for example: 

Here, NumPy will generate a list of equally distant values from -3 to 3.

```python

x_axis = np.linspace(-3,3)
x_axis
```

We can use these values as "seed" or x parameters to obtain the Y values using any linear algebra function. This is an example using the square function.

```python
x_axis = np.linspace(-3,3)

def square(x): return x**2

y_axis = square(x_axis)

plt.plot(x_axis, y_axis)
```

## The Parent functions in linear algebra

Data patterns can be described by algebraic functions, and you can plot and visualize them in charts.

The following are called the "parent functions" in linear algebra, and they will usually cover 95% of your data patterns.

![Parent functions](https://breathecode.herokuapp.com/v1/media/file/parent-functions-jpg?raw=true)

## Plotting the parent functions

Here are some examples of plotting the parent functions in linear algebra. First we import NumPy because it contains the functions:


```python
import numpy as np

# Start by defining a simple X axis and the Y axis will be given by our parent function
x_axis = np.linspace(-3,3,1000)
```

### Square Function 

```python
def square(x): return x**2
plt.plot(x_axis, square(x_axis))
```

### Sen Function 

```python
def sen_function(x): return np.sin(x)
plt.plot(x_axis, sen_function(x_axis))
```

### Exponential Function

```python
def exponential(x): return 100*(np.power(2, x))

plt.plot(x_axis, exponential(x_axis))
```

### Logarithmic Function

```python
def logarithmic(x): return np.log10(x)

plt.plot(x_axis, logarithmic(x_axis))
```
